```python
import pandas as pd
import os
from tkinter import Tk, filedialog
root = Tk() # pointing root to Tk() to use it as Tk() in program.
root.withdraw() # Hides small tkinter window.
root.attributes('-topmost', True) # Opened windows will be active. above all windows despite of selection.
```

```python
direct = filedialog.askdirectory() # Returns opened path as str
```

```python
direct
```

```python
filenames = list(filter(lambda x: 'Results.csv' in x, os.listdir(direct)))
srow=list(range(2,104))
```

```python
filenames
```

```python
summary=[]
for filename in filenames:
    res = (
        pd.read_csv(f'{direct}/{filename}',
            sep='\t', encoding='cp852',
            index_col=0, header=[0,1],
            skiprows=srow)
    )
    sn = filename[3:23]
    pulse = int(res['ClTime1'].loc['Average']-res['OpTime1'].loc['Average'])
    qAvg = res['Q1','[mg]'].loc['Average']
    qMin = res['Q1','[mg]'].loc['Min']
    qMax = res['Q1','[mg]'].loc['Max']
    qStdDev = res['Q1','[mg]'].loc['Stand.Dev']
    od = res['OpDelay1', '[us]'].loc['Average']
    cd = res['ClDelay1', '[us]'].loc['Average']
    density=res['Density chamber','[mg/mm^3]'].loc['Average']
    
    if res['Q1','[mg]'].loc['Stand.Dev'] > 0:
        cov = res['Q1','[mg]'].loc['Stand.Dev']/res['Q1','[mg]'].loc['Average'] * 100
    else:
        cov = 0
        
    list = [sn, pulse, qAvg, qMin, qMax, qStdDev, od, cd, density, cov]
    summary.append(list)
```

```python
summary
```

```python
col =(
    ['SN','Pulse [us]','Q Avg [mg/shot]','Q min [mg/shot]','Q max [mg/shot]',
     'Q std.dev [mg/shot]', 'ODL [us]','CDL [us]','Density [g/cm3]','COV Shot-to-Shot [%]']
)
```

```python
col
```

```python
pd.DataFrame(summary,columns=col)
```

```python
pd.DataFrame(summary,columns=col).to_excel(f'{direct}/summary.xlsx', sheet_name='InjQty' , index=False)
```
